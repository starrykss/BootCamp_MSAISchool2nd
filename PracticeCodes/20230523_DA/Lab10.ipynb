# Lab 10. T 분포 간단 실습
---

```python
import numpy as np
import matplotlib.pyplot as plt
```

- 데이터를 생성하고 간단하게 시뮬레이션을 해본다.

```python
df = 10     # 자유도 
num_simulations = 1000 
t_values = np.random.standard_t(df, num_simulations)

print(t_values)
```

- 간단하게 시각화를 진행해본다.

```python
# 시각화 
plt.hist(t_values, bins=20, alpha=0.5)
plt.show()
```
