# Lab 08. 베타 분포 모형 간단 실습
---

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
# 모수 설정 
alpha = 1
beta = 10 

user_data = np.random.beta(alpha, beta, size=1000)

print(user_data)
```

- 사용자 행동 데이터의 분포를 시각화 해본다.

```python
plt.hist(user_data, bins=10, density=True, alpha=0.8)

plt.show()
```
