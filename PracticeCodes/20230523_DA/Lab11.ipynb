# Lab 11. F 분포 간단 실습
---

```python
import numpy as np
import matplotlib.pyplot as plt 
```

- F 분포를 모델링하고 시뮬레이션 해본다.

```python
# F 분포 모델링 
dfn = 5
dfd = 10 

# F 분포 시물레이션 
num_simulations = 1000 
f_values = np.random.f(dfn, dfd, num_simulations)

print(f_values)
```

- 간단하게 시각화를 진행해본다.

```python
plt.hist(f_values, bins=20, alpha=0.5)
```
