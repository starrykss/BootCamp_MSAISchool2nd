# Lab 09. 카이 제곱 모형 간단 실습
---

```python
import numpy as np
import matplotlib.pyplot as plt
```

- 카이 제곱 분포에서 랜덤 변수를 생성해본다.

```python
# 모순 설정 
df = 5     # (자유도)

# 카이 제곱 분포에서 랜덤 변수 생성 
random_variable = np.random.chisquare(df, size=1000)

print(random_variable)
```

- 간단하게 시각화를 해본다.

```python
plt.hist(random_variable, bins=30, density=True, alpha=0.8)
plt.show()
```
