# Lab 01. 인공 신경망 : 가중치 적용한 시각화 실습
---

```python
import torch
import matplotlib.pyplot as plt
```

```python
# 2차원 가중치 행렬을 랜덤으로 초기화 
w = torch.randn(3, 3)
print(w)

w_flattend = w.flatten()
print(w_flattend)
```

- 시각화를 해본다.

```python
plt.hist(w_flattend, bins=20)
plt.xlabel('Weight value')
plt.ylabel('Frequency')

plt.show()
```

## 크기가 1인 랜덤한 값을 가지는 편향 생성하기

```python
b = torch.rand(1)    # 편향을 랜덤으로 초기화 

print(b)

# 편향 값을 시각화 
plt.hist(b.detach().numpy(), bins=20)     # b.detach().numpy() -> 텐서를 넘파이 배열로 변환 
plt.xlabel('Bias value')
plt.ylabel('Frequency')

plt.show()
```
