# Lab 05. 소리 데이터에 적합한 데이터 증강 기법 - Stretching
---

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import soundfile as sf 
```

```python
# 1. 샘플 데이터 불러오기 
data, sr = librosa.load('./data/blues.00001.wav')
```

```python
# 2. 오디오 신호를 스트레칭 
data_stretch = librosa.effects.time_stretch(data, rate=0.8)
```

```python
# 3. 결과 저장 
save_path = "./data/sample_stretech.wav"
sf.write(save_path, data_stretch, sr)
```

## 원본 데이터와 늘린 데이터의 비교 

```python
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
librosa.display.waveshow(data, sr=sr, alpha=0.8)

plt.subplot(1, 2, 2)
librosa.display.waveshow(data_stretch, sr=sr, alpha=0.8)

plt.title("Stretch")
plt.show()
```
