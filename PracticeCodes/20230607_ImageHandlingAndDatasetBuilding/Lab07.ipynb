```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = "./data/box.png"
```

```python
image_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

image_10x10 = cv2.resize(image_gray, (10, 10))    # 이미지를 10x10 픽셀 크기로 변환한다. 
image_10x10.flatten()

plt.imshow(image_10x10, 'gray')
plt.show()
```

```python
image_color = cv2.imread("./data/iamge01.png")
image_color = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)


image_10x10_temp = cv2.resize(image_color, (10, 10))    # 이미지를 10x10 픽셀 크기로 변환한다. 
image_10x10_temp.flatten()

plt.imshow(image_10x10_temp)
plt.show()
```
