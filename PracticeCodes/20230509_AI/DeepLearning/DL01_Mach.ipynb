## 1차 함수
---

```python
def linear_function(x):
    a = 0.5
    b = 2

    return a*x + b    # f(x)=ax+b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5, 5, 0.1)     # -5부터 +5까지 0.5씩 증가]
y = linear_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

## 2차 함수
---

```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2

    return a*x**2 + b*x + c    # y=ax^2+bx+c
```

```python
print(quadratic_function(2))
```

```python
x = np.arange(-5, 5, 0.1)
y = quadratic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

## 3차 함수
---

```python
def cubic_function(x):
  a = 4
  b = 0
  c = -1
  d = -8

  return a*x**3 + b*x**2 + c*x + d    # y=ax^3+bx^2+cx+d
```

```python
x = np.arange(-5, 5, 0.1)
y = cubic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cublic Function')
```

### 1차원 텐서
---

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

# 차원이 달라도 연산이 된다.
c = a + b     # 1차원 텐서 사이의 연산
d = a + x     # 1차원과 0차원 텐서 사이의 연산

for i in [a, b, c, d]:
    print("값:", i)
    print("형태:", i.shape)
    print(i.ndim, "차원", '\n')
```

- `dot()` 연산은 2개의 벡터가 있을 때, 각각의 성분의 곱을 모두 더하는 연산이다.

```python
x = np.array([1, 2, 0])
y = np.array([0, 2, 1])
z = np.dot(x, y)    # 각각의 벡터에 있는 성분의 곱을 모두 더한다.

print("값:", z)
print("형태:", z.shape)
print(z.ndim, "차원")
```

### ① AND
---

```python
def AND(a, b):
  input = np.array([a, b])
  weights = np.array([0.4, 0.4])
  bias = -0.6
  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

```python
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)

bias = -0.6

y = (-0.4 * x1 - bias) / 0.4

plt.axvline(x = 0)
plt.axhline(y = 0)

plt.plot(x1, y, 'r--')
plt.scatter(0, 0, color='orange', marker='o', s=150)
plt.scatter(0, 1, color='orange', marker='o', s=150)
plt.scatter(1, 0, color='orange', marker='o', s=150)
plt.scatter(1, 1, color='black', marker='^', s=150)

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.grid()
```

### ② OR
--- 

```python
def OR(a, b):
  input = np.array([a, b])
  weights = np.array([0.4, 0.4])
  bias = -0.3
  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0, 0))
print(OR(0, 1))
print(OR(1, 0))
print(OR(1, 1))
```

```python
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)

bias = -0.3

y = (-0.4 * x1 - bias) / 0.4

plt.axvline(x = 0)
plt.axhline(y = 0)

plt.plot(x1, y, 'r--')
plt.scatter(0, 0, color='orange', marker='o', s=150)
plt.scatter(0, 1, color='black', marker='^', s=150)
plt.scatter(1, 0, color='black', marker='^', s=150)
plt.scatter(1, 1, color='black', marker='^', s=150)

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.grid()
```

### ③ NAND
--- 

```python
def NAND(a, b):
  input = np.array([a, b])
  weights = np.array([-0.5, -0.5])
  bias = 0.7
  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(NAND(0, 0))
print(NAND(0, 1))
print(NAND(1, 0))
print(NAND(1, 1))
```

```python
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)

bias = 0.7

y = (0.5 * x1 - bias) / -0.5

plt.axvline(x = 0)
plt.axhline(y = 0)

plt.plot(x1, y, 'r--')
plt.scatter(0, 0, color='black', marker='o', s=150)
plt.scatter(0, 1, color='black', marker='o', s=150)
plt.scatter(1, 0, color='black', marker='o', s=150)
plt.scatter(1, 1, color='orange', marker='^', s=150)
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.grid()
plt.show()
```

### ④ XOR
--- 

- `AND`, `OR`, `NAND`처럼 하나의 퍼셉트론으로 구현 불가능
- 여러 개의 퍼셉트론을 사용하여 구현

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)

  return y
```

```python
print(XOR(0, 0))
print(XOR(0, 1))
print(XOR(1, 0))
print(XOR(1, 1))
```

### 계단 함수(Step Function)
---

```python
def step_function(x):
  if x > 0:
    return 1
  else:
    return 0
```

```python
print(step_function(0))
print(step_function(-1))
print(step_function(0.1))
```

```python
def step_function_for_numpy(x):
  y = x > 0
  return y.astype(np.int_)
```

```python
import numpy as np

a = np.array([5, 3, -4, 2, 0])
print(step_function_for_numpy(a))
```

### Sigmoid Function 
---

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
print(sigmoid(3))
print(sigmoid(10))
print(sigmoid(100))
print(sigmoid(-3))
print(sigmoid(-100))
```

#### Step Function vs. Sigmoid Function
---

```python
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.01)
y1 = sigmoid(x)
y2 = step_function_for_numpy(x)

plt.plot(x, y1, 'r-', x, y2, 'b--')
```

### Softmax Function
---

```python
def softmax(x):
  exp_x = np.exp(x)
  sum_exp_x = np.sum(exp_x)
  y = exp_x / sum_exp_x

  return y
```

```python
x = np.array([0.3, 0.2, 3.0, -1, 2])
print(softmax(x))
print(np.sum(softmax(x)))
```
