# Lab 09. 시계열 데이터에서 결측치를 평균값으로 대체한 후 시각화 해보기
---

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

- 시계열 데이터를 생성하고 결측치를 생성한다. 그리고 결측치를 확인해본다.

```python
# 시계열 데이터 생성 
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)

# 데이터 프레임 생성
data = pd.DataFrame({
    'Date' : dates,
    'Value' : values
})

print(data)
# 결측치 생성 
data.loc[50:100, 'Value'] = np.nan

print(data.loc[50:100])
```

- 결측치를 평균값으로 대체해본다.

```python
mean_value = data['Value'].mean()
print(mean_value)

data['Value_filled'] = data['Value'].fillna(mean_value)
print(data.loc[50:100])
```

- 시각화를 해본다.

```python
# 시계열 데이터 시각화
plt.plot(data['Date'], data['Value'], 'o-', label='Original')
plt.plot(data['Date'], data['Value_filled'], 'x-', label='Mean Imputation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Mean Imputation')
plt.legend()

plt.show()
```
