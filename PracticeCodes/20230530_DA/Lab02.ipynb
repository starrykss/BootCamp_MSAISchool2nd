# Lab 02. 시계열 데이터 탐색 및 시각화 실습

- 시계열 데이터 시각화 기법을 알아보자. (선 그래프, 박스 플롯, 자기 상관 그래프)

````python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
````

- 시계열 데이터를 생성해본다.

````python
# 시계열 데이터 생성 
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)

# 데이터 프레임 생성 
data = pd.DataFrame({
    'Date': dates, 
    'Value' : values
})
````

- 시계열 데이터를 플롯으로 시각화 해본다.

````python
# 시계열 데이터 시각화 
plt.plot(data['Date'], data['Value'])
plt.xticks(rotation=-45)  

plt.show()
````

## 산점도 그래프 시각화

- 시계열 데이터를 생성하고 확인해본다.

````python
# 시계열 데이터 생성 
date_temp = pd.date_range(start='2023-01-01', periods=1000, freq='D')
val_temp = np.random.randn(1000)

# 프레임 생성
data_pd = pd.DataFrame({
    'Date' : date_temp, 
    'Vale' : val_temp
})

print(data_pd)
````

- 산점도 그래프로 시각화 해본다.

````python
plt.scatter(data_pd['Date'], data_pd['Vale'])
plt.xticks(rotation=-45)  

plt.show()
````

## 막대 그래프 시각화 

- 시계열 데이터를 생성하고 확인해본다.

````python
dates_box_plot = pd.date_range(start='2023-01-01', periods=1000, freq='D')
val_box_plot = np.random.randn(1000)

box_plot_data = pd.DataFrame({
    'Date': dates_box_plot, 
    'Value': val_box_plot
})

print(box_plot_data)
````

- 막대 그래프로 시각화 해본다.

````python
plt.bar(box_plot_data['Date'], box_plot_data['Value'])
plt.xticks(rotation=-45)  

plt.show()
````

## 박스 플롯 시각화

- 시계열 데이터를 생성하고 확인해본다.

````python
box_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
box_val = np.random.randn(1000)

box_pd = pd.DataFrame({
    'Date': box_date,
    'Value' : box_val
})

print(box_pd)
````

- 박스 플롯으로 시각화 해본다.

## 히스토그램 시각화

- 시계열 데이터를 생성하고 확인해본다.

````python
histogram_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
histogram_val = np.random.randn(1000)

histogram_data = pd.DataFrame({
    'Date' : histogram_date,
    'Value' : histogram_val
})

print(histogram_data)
````

## 히트맵 시각화 

- 시계열 데이터를 생성하고 확인해본다.

````python
import seaborn as sns

hit_date = pd.date_range(start='2023-01-01', periods=1000, freq="D")
hit_val = np.random.randn(1000)

hit_data = pd.DataFrame({
    'Date' : hit_date,
    'Value' : hit_val
})

print(hit_data)
````

- 데이터를 피벗 테이블로 변환한다.
    - 피벗 테이블은 <ins>데이터를 특정 열을 기준으로 그룹화하여 요약</ins>하는 방법이다.

예)
```
Date      value 
23-01-01    A
23-01-01    B
23-01-02    A
23-01-02    C
23-01-03    B
23-01-03    C
```

피벗 테이블 생성 
```
          A   B   C 
23-01-01  A   B  NAN
23-01-02  A  NAN  C
23-01-03 NAN  B   C 
```

````python
# 데이터를 피벗 테이블 변환 
pivot_table = data.pivot(index='Date', columns='Value', values='Value')

print(pivot_table)
````

- 히트맵으로 시각화해본다.

````python
# 히트맵 시각화 
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, cmap='YlOrRd', cbar=True)

plt.show()
````
