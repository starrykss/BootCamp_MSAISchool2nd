# Lab 06. pandas를 활용한 선형 보간 실습
---

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

- 시계열 데이터를 생성하고 확인해본다.

```python
# 시계열 데이터 생성 
dates = pd.date_range(start='2023-01-01', periods=10, freq='D')
values = [1, np.nan, 3, np.nan, 5, 6, np.nan, 8, 9 , np.nan]

# 데이터 프레임 생성 
data = pd.DataFrame({
    'Date' : dates, 
    'Value' : values
})

print(data)
```

- 선행 보간을 적용해본다.

```python
data['Value_linear'] = data['Value'].interpolate(method='linear')

print(data)
```

- 시각화 해본다.

```python
plt.plot(data['Date'], data['Value'], 'o-', label = 'original data')
plt.plot(data['Date'], data['Value_linear'], 'x-', label = 'Linear Interpolation')
plt.legend()
plt.xticks(rotation=-45)

plt.show()
```
