```python
import numpy as np 
import pandas as pd 
```

- 가상의 데이터를 생성한다. 이 데이터는 1학년부터 3학년까지 학생들의 수학, 영어, 국어 점수를 나타내며, 각 학년당 100명의 학생이 있다고 가정한다.

```python
# 가상 데이터 생성 -> 학생 데이터
np.random.seed(42)
math_scores = np.random.normal(loc=70, scale=10, size=(300,))
english_scores = np.random.normal(loc=80, scale=8, size=(300,))
korean_scores = np.random.normal(loc=65, scale=12, size=(300,))

# 학년 데이터 생성 
grades = np.repeat([1, 2, 3], 100)

# 데이터 프레임 생성 
df = pd.DataFrame({
    'Grade' : grades,
    'Math' : math_scores,
    'English' : english_scores,
    'Korean' : korean_scores
})
```

- `.csv` 파일로 저장하고 읽어온다.

```python
# csv 파일로 저장 
df.to_csv('./data/student_scores.csv', index=False, encoding='utf-8')
```

```python
# csv 읽기 
df = pd.read_csv('./data/student_scores.csv')
print(df)
```

- 상관 관계를 분석하고 히트맵으로 시각화 해본다.

```python
# 상관 분석 : corr()
df_temp = df.corr()

print(df)
```

```python
# 상관 분석 히트맵 시각화 
import seaborn as sns

sns.heatmap(df_temp , annot=True)
```

## **각 과목별로 성적 분포와 통계치 분석하기**
--- 

- 과목별로 성적 분포와 통계치를 분석해본다. (판다스의 `describe()` 함수를 사용한다.)

```python
# 수학 성적 분포 분석 
math_stats = df['Math'].describe()

# 영어 성적 분포 분석
english_stats = df['English'].describe()

# 국어 성적 분포 분석 
korean_stats = df['Korean'].describe()

print(korean_stats)
```

- 과목별로 성적 분포와 통계치를 분석하고 시각화 해본다.

```python
import matplotlib.pyplot as plt

# 수학 성적 분포 시각화
plt.hist(df['Math'], bins=10)
plt.title('Math Scores Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

# 영어 성적 분포 시각화
plt.hist(df['English'], bins=10)
plt.title('English Scores Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

# 국어 성적 분포 시각화
plt.hist(df['Korean'], bins=10)
plt.title('Korean Scores Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```
