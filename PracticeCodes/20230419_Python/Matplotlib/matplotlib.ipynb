## **그래프 색상 지정하기**
---

```python
plt.plot(x, np.sin(x), color='blue')         # 그래프 색상 지정 (파란색)
plt.plot(x, np.sin(x - 1), color='g')        # 약어를 사용하여 그래프 색상 지정 (녹색)
plt.plot(x, np.sin(x - 2), color='k')        # 약어를 사용하여 그래프 색상 지정 (검정색)
plt.plot(x, np.sin(x - 3), color='0.75')     # 0-1 사이 색상, 1 : 검정색, 0 : 흰색 (그레이 스케일로 표시)
plt.plot(x, np.sin(x - 4), color='#FFDD44')        # 색상 코드를 사용하여 그래프 색상 지정
plt.plot(x, np.sin(x - 5), color=(1.0, 0.2, 0.3))  # R: 1.0, G: 0.2, B: 0.3
plt.plot(x, np.sin(x - 5), color='chartreuse')     # 색상 이름 사용하기
```

## **그래프 모양 지정하기**
---

### **용어 사용하기**
---

```python
plt.plot(x, x + 0, linestyle="solid")
plt.plot(x, x + 1, linestyle="dashed")
plt.plot(x, x + 2, linestyle="dashdot")
plt.plot(x, x + 3, linestyle="dotted")
```

### **기호 사용하기**
---

```python
plt.plot(x, x + 0, linestyle="-")          # solid
plt.plot(x, x + 1, linestyle="--")         # dashed
plt.plot(x, x + 2, linestyle="-.")         # dashdot
plt.plot(x, x + 3, linestyle=":")          # dotted
```

### **기호를 사용하여 그래프 색상과 모양 함께 지정하기**
---

```python
plt.plot(x, x + 0, '-g')       # solid green
plt.plot(x, x + 1, '--c')      # dashed cyan
plt.plot(x, x + 2, '-.k')      # dashdot green
plt.plot(x, x + 3, ':r')       # dotted red
```

## **세부 조정하는 법**
---

```python
plt.plot(x, np.sin(x))   # 아무 옵션을 주지 않으면, 항상 그래프를 꽉 채워서 표현하려는 성향이 있다.

# Limit 조정하기
plt.xlim(-1, 11)         # x축의 범위를 -1 ~ 11로 지정
plt.ylim(-1.5, 1.5)      # y축의 범위를 -1.5 ~ 1.5로 지정
```

```python
# 그래프의 좌우, 상하 반전
plt.plot(x, np.sin(x))
plt.xlim(10, 0)          # X축의 좌우를 바꾸어 출력한다.
plt.ylim(1.2, -1.2)      # y축의 상하를 바꾸어 출력한다.
```

```python
# x축, y축 범위를 한꺼번에 지정하기
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])      # 축을 리스트 타입으로 설정해준다. [x축_x, x축_y, y축_x, y축_y]
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')     # 그래프를 타이트하게 출력한다. (기본값, 실제 그래프 사이즈가 아니다.)
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')         # x, y 비율을 동일하게 해준다. (✅ 실제 그래프의 모양이다.) 
```

## **산점도 출력하기**
---

- **산점도(Scatter Plot)**
- 머신러닝을 할 때 많이 사용한다.

### **산점도의 표현**
---

```python
plt.plot(x, np.sin(x), 'o')     # o 모양의 marker를 표시한다.
```

```python
x = np.linspace(0, 10, 30)     # 30개만 만든다.
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```python
rng = np.random.RandomState(0)     # 0 : 시드(Seed), 다른 사람이 똑같이 0을 넣을 경우, 다른 사람도 나와 똑같은 랜덤값을 얻을 수 있다.
rng.rand(5)
```

```python
plt.plot(rng.rand(5), rng.rand(5), 'o')       # x 좌표는 랜덤으로 5개, y 좌표는 랜덤으로 5개씩 표시한다.
```

```python
# 스타일 지정하기
plt.style.use('bmh')

# 마커의 종류
rng = np.random.RandomState(0)
for marker in ['o', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd', 'p']:                            # ',' 마커는 먼지처럼 작게 표시 된다.
    plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
    plt.legend(numpoints=1)      # 포인트의 개수를 1개만 출력 (기본값: 2개)
    plt.xlim(0, 1.8)             # 범례에 의해 가려지는 마커가 없도록 x축 길이 조정
```

```python
# 스타일 원래대로 돌리기
plt.style.use('classic')
```

```python
plt.plot(x- 2, y - 2)                    # 산점도의 그래프에 선을 연결하기
plt.plot(x - 1, y - 1, 'o')              # 산점도로 표시하려면 marker를 추가한다.
plt.plot(x, y, '-o')                     # 산점도로 표시하고 선을 연결해서 표현한다.
plt.plot(x + 1, y + 1, '--ok')           # 산점도로 표시하고 선을 연결한 후 색깔을 넣어 표현한다.
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15,             # marker 사이즈 변경 
         linewidth=4,               # 선의 굵기 변경
         markerfacecolor='white',   # marker 색상을 흰색으로 변경
         markeredgecolor='red',     # marker의 테두리 색 변경
         markeredgewidth=2          # marker의 굵기 변경
)
```

## `plt.scatter`를 사용하여 산점도 표현하기
---

```python
rng = np.random.RandomState(0)   # rng : random number generation(generator)의 약자
x = rng.randn(100)                # 랜덤값 100개 만들기
y = rng.randn(100)                

# rand(x) : 0 ~ 1 사이의 값들을 x개 뽑아낸다.
# randn(x) : 0을 기준으로 해서 표준편차가 1인 양수/음수 값들을 x개 뽑아낸다.
# randint(a, b) : a<=x<b 사이의 임의의 정수형 값 1개를 뽑아낸다.
## np.random.randint()와 같이 사용할 경우, 실행시킬 때마다 다른 값이 나온다.

sizes = rng.rand(100) * 1000     # 랜덤으로 사이즈를 정한다.
color = rng.rand(100)            # 랜덤으로 색상을 정한다.

plt.scatter(x, y, s=sizes, c=color, alpha=0.3, cmap='viridis')     # alpha는 투명도를 지정한다. 컬러맵을 적용하면 일관성 있는 색상을 출력할 수 있다.
```

## **오차 시각화**
---

- 100% 정확한 AI는 없다. (100% 정확하다는 것은 의미가 없다.)
    - 예측하는 것이 의미가 없다.

```python
x = np.linspace(0, 10, 50)
dy = 0.8    # 에러 범위 지정
y = np.sin(x + np.random.randn(50))

plt.errorbar(x, y, yerr=dy, fmt='.k')    # y축에 대한 에러값 범위를 dy로 설정

# 이 값은 y축의 이 범위에서 오차가 있을 수 있다.
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3)
```
