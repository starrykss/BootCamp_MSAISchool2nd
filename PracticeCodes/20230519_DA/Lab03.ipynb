# Lab 03. 중심 경향성 간단한 시각화 실습
---

- 평균(mean), 중앙값(median), 최빈값(mode)을 구하고 이를 시각화 해보자.

```python
import numpy as np
import matplotlib.pyplot as plt
```

- 데이터를 생성하고 확인해본다.

```python
# 데이터 생성
data = np.random.normal(loc=0, scale=1, size=10000)

print(data)
```

- 중심 경향성을 계산해보고 출력해본다.

```python
# 중심 경향성 계산
# 평균(mean), 중앙값(median), 최빈값(mode)
mean = np.mean(data)
median = np.median(data)
mode = np.round(np.mean(data))

print(mean, median, mode)
```

- 분포를 시각화 해본다.

```python
# 분포 시각화
fig, ax = plt.subplots()
ax.hist(data, bins=50)
ax.axvline(mean, color='red', linestyle='dashed', linewidth=2)
ax.axvline(median, color='green', linestyle='dashed', linewidth=2)
ax.axvline(mode, color='purple', linestyle='dashed', linewidth=2)
ax.legend(['Mean: {:.2f}'.format(mean), 'Median: {:.2f}'.format(median), 'Mode: {}'.format(mode)])
ax.set_xlabel('Data Values')
ax.set_ylabel('Frequency')
ax.set_title('Data Distribution')

plt.show()
```

> 먼저 데이터는 정규 분포를 따르며, x축은 데이터 값, y축은 빈도수(Frequency)를 나타낸다. 분포는 평균을 중심으로 좌우 대칭인 종모양의 형태를 갖는다. 빨간색, 초록색, 보라색 점선은 각각 평균, 중앙값, 최빈값을 나타낸다. 평균은 데이터의 전체 합을 데이터 개수로 나눈 값으로, 그래프에서 분포의 중심을 대표한다. 중앙값은 데이터를 크기순으로 정렬했을 때 가운데 위치한 값으로, 이상치(Outlier)에 영향을 받지 않는 강건한 지표이다. 최빈값은 가장 자주 나타나는 값으로, 이 그래프에서는 약 `-0.0` 정도에 위치해 있다.
